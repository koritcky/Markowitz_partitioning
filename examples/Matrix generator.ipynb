```python
import numpy as np
import time
import os
os.chdir('/home/koritskiy/Yandex.Disk/personal/rqc/markowitz_partitioning')
os.getcwd()
from modules.partitioning import Partitioning
```

```python
import itertools
block_dims = []
for length in range(3, 7):
    for dims in itertools.combinations_with_replacement([3,4,5,6], length):
        block_dims.append(list(dims))
print(len(block_dims))
```

```python
for block_dim in block_dims:
    block_dim_name = ''.join([str(i) for i in block_dim])
    block_mat = Partitioning.rand_sym_block_gen(block_dim, order = True)
    (_, mixed_mat, mix_permutation_mat) = Partitioning.mixed_matrix_generator(block_mat=block_mat)
    path = "/home/koritskiy/Yandex.Disk/personal/rqc/markowitz_partitioning/data/test_matrices/" + str(block_dim_name)
    if not os.path.exists(path):
        os.makedirs(path)
    np.savetxt(path + "/block_mat" + str(block_dim_name) + ".csv", block_mat, delimiter=",")
    np.savetxt(path + "/mixed_mat" + str(block_dim_name) + ".csv", mixed_mat, delimiter=",")
    np.savetxt(path + "/mix_permutation_mat" + str(block_dim_name) + ".csv", mix_permutation_mat, delimiter=",")
```
